```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import urllib
import urllib.request
from urllib.request import urlopen
import json
import xmltodict
from tqdm import tqdm
from lxml import html
import requests, bs4


def return_fcst_info(date, time, pageNo):
    key='zkwKOhULzACqZDpQze92X07T7ZKmqj+UPdYt29/2NT7ujTCl4MaGwBOk2SOVQTaRdOUWpogDkwOrQnwGqx31cg=='
    url = 'http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getVilageFcst'
    queryParams = '?' + urllib.parse.urlencode(
        {
            urllib.parse.quote_plus('ServiceKey') : key, 
            urllib.parse.quote_plus('pageNo') : pageNo, 
            urllib.parse.quote_plus('numOfRows') : '1000', 
            urllib.parse.quote_plus('dataType') : 'xml', 
            urllib.parse.quote_plus('base_date'): date,
            urllib.parse.quote_plus('base_time') : time,
            urllib.parse.quote_plus('nx') : '65',
            urllib.parse.quote_plus('ny') : '103'
        }
    )

    response = requests.get(url + queryParams).text.encode('utf-8')
    xmlobj = bs4.BeautifulSoup(response, 'lxml-xml')
    rows=xmlobj.findAll('item')
    return rows
```

```python
rows=return_fcst_info("20211027","0800","1")

rowList = []
nameList = []
columnList = []

rowsLen = len(rows)
for i in range(0, rowsLen):
    columns = rows[i].find_all()

    columnsLen = len(columns)
    for j in range(0,columnsLen):

        if i==0:
            nameList.append(columns[j].name)
        eachColumn = columns[j].text
        columnList.append(eachColumn)
    rowList.append(columnList)
    columnList=[]    
result = pd.DataFrame(rowList, columns=nameList)
```
